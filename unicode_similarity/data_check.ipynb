```python
import glob
import os

from PIL import Image
import numpy as np
import cv2
```

```python
image_paths = glob.glob('./output/*.png')
image_paths
```

```python
missed_image = np.array(Image.open('./output/1b3a.png'))
```

```python
missed_image_paths = []

for image_path in image_paths:
    image = np.array(Image.open(image_path))
    if np.allclose(missed_image, image):
        missed_image_paths.append(image_path)
        
missed_image_paths
```

```python
for missed_image_path in missed_image_paths:
    name = missed_image_path.split('/')[2]
    os.rename(missed_image_path, './missed/' + name)
```

```python
invalid_image = cv2.imread('./0a00.png')
```

```python
invalid_image_paths = []

for image_path in image_paths:
    image = cv2.imread(image_path)
    res = cv2.matchTemplate(image, invalid_image, cv2.TM_CCORR_NORMED)
    minVal, maxVal, minLoc, maxLoc = cv2.minMaxLoc(res)
    
    if maxVal > 0.8:
        invalid_image_paths.append(image_path)

invalid_image_paths
```

```python
for invalid_image_path in invalid_image_paths:
    name = invalid_image_path.split('/')[2]
    os.rename(invalid_image_path, './invalid/' + name)
```
