```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('vector.csv', index_col=0)
df
```

```python
available_codes = frozenset(df['Code'])
available_codes
```

```python
def cos_sims(char_vec):
    res = []
    for idx, row in df.iterrows():
        v1 = row.iloc[0:16]
        cos_sim = np.dot(v1, char_vec) / (np.linalg.norm(v1) * np.linalg.norm(char_vec))
        res.append([cos_sim, row['Code']])
    return res
```

```python
def search(code):
    search_vec = np.array(df[df['Code'] == code].iloc[:, 0:16])[0]
    res = cos_sims(search_vec)
    res.sort(reverse=True)

    print('%s に類似する見た目の文字：' % chr(int(code, base=16)))

    for (score, candidate) in res[:50]:
        print('%s: %s  (%.6f accuracy)' % (candidate, chr(int('0x' + candidate, base=16)), score))
```

```python
search('0041')
```

```python
import ngtpy
```

```python
vec = np.array(df.iloc[:, 0:16])

ngtpy.create(b'index', dimension=vec.shape[1], distance_type="Cosine")
index = ngtpy.Index(b'index')
index.batch_insert(vec)
index.build_index()
index.save()
```

```python
def search_by_ngt(code):
    search_vec = np.array(df[df['Code'] == code].iloc[:, 0:16])[0]
    res = index.search(search_vec, size=50)
    print(res[0])
#     res.sort(reverse=True)

    print('%s に類似する見た目の文字：' % chr(int(code, base=16)))

    for (idx, score) in res[:50]:
        candidate = df.iloc[idx, 16]
        print('%s: %s  (%.6f dist)' % (candidate, chr(int('0x' + candidate, base=16)), score))
```

```python
search_by_ngt('004c')
```
